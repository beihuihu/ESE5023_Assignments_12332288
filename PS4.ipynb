```python
import numpy as np
import xarray as xr
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
import matplotlib.ticker as ticker
%matplotlib inline
```

1. Global Earthquakes

```python
Eqs=pd.read_csv("usgs_earthquakes.csv")
```

```python
Eqs_50=Eqs.sort_values(by='mag',ascending=False).head(50)
Eqs_50
```

```python
# Create and define the size of a figure object 
plt.figure(figsize=(10,4), dpi=100)
    
# Create an axes with Orthographic projection style
ax = plt.axes(projection=ccrs.Robinson(central_longitude=180))

# Add natural features to axes using cartopy.feature (cfeature)
ax.stock_img()
plt.scatter('longitude', 'latitude', data=Eqs_50, 
             c='mag', cmap='Reds', 
            edgecolors='black', linewidths=0.3,transform=ccrs.PlateCarree())
cb1=plt.colorbar(fraction=0.01)

tick_locator = ticker.MaxNLocator(nbins=9)  # colorbar
cb1.locator = tick_locator
cb1.update_ticks()
cb1.set_label('magnitude', x=1.1, y=0.5)
plt.title('Top 50 earthquakes of 2014')
```

2. Explore a netCDF dataset

```python
ds=xr.open_dataset('GLDAS_CLSM025_DA1_D.A20230630.022.nc4.SUB.nc4', engine="netcdf4")
GWS=ds.GWS_tavg.isel(time=-1)
GWS
```

```python
plt.figure(figsize=(10,4), dpi=100)
#set projection
proj = ccrs.PlateCarree() 
ax   = plt.axes(projection=proj)

#add  feature, colorbar
ax.add_feature(cfeature.OCEAN, zorder=0)
ax.coastlines() 
GWS.plot.contourf(ax=ax,transform=ccrs.PlateCarree(),
               vmin=77.02637, vmax=3578.3328, levels=11, cmap='Blues',
               add_colorbar=True)

# set gridlines, label and ticks of axis ,title
gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,linewidth=1, color='black', alpha=0.5)
gl.ylocator = mticker.FixedLocator(np.arange(-90,90,30))
gl.xlocator = mticker.FixedLocator(np.arange(-180, 180, 30))
gl.top_labels = False
gl.right_labels = False
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER
gl.xlabel_style = {'size': 10, 'color': 'gray'}
gl.ylabel_style = {'size': 10, 'color': 'gray'}
plt.title('global mean ground water storage of 2023-06-30',fontsize=15)

#add legend
Beijing = dict(lon=115.25, lat=39.28)
Moskva = dict(lon=37.54, lat=55.67)
lons = [Beijing['lon'], Moskva['lon']]
lats = [Beijing['lat'], Moskva['lat']]
ax.plot(lons, lats, 'go',label='city', transform=ccrs.PlateCarree())
ax.legend()

#add annotation and text box
plt.annotate('Beijing', xy=(115.25, 39.28), xytext=(145, 20),
             bbox=dict(boxstyle='square', fc='white', linewidth=0.1),
             arrowprops=dict(facecolor='blue', shrink=0.01, width=0.1,headlength=2,headwidth=2),
             fontsize=10, color='black', horizontalalignment='center')
plt.annotate('Moskva', xy=(37.54,55.67 ), xytext=(0, 75),
             bbox=dict(boxstyle='square', fc='white', linewidth=0.1),
             arrowprops=dict(facecolor='blue', shrink=0.01, width=0.1,headlength=2,headwidth=2),
             fontsize=10, color='black', horizontalalignment='center')
text='prijection: Plate Carree'
plt.text(-175, -50, text, fontsize=10,
         bbox=dict(boxstyle='round', facecolor='white', edgecolor='gray', alpha=0.7))
```
